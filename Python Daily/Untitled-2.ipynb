```python
import numpy as np

first = np.arange(1, 21).reshape(4, 5)

# ماسک مضرب‌های 3 از خود آرایه‌ی دو بعدی
maskbool = first % 3 == 0

# اعمال ماسک
result = first[maskbool]

print(result)  # خروجی: [ 3  6  9 12 15 18]
```
